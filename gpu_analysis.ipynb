# Projeto Final – Análise de Benchmark de GPUs
Este notebook realiza a análise exploratória e tratamento dos dados do dataset **gpu_benchmark_60.csv**, que servirá como base para o dashboard final em Streamlit.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Carregando o Dataset

```python
df = pd.read_csv('gpu_benchmark_60.csv')
df.head()
```

## Informações gerais sobre os dados

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

## Tratamento e Limpeza dos Dados

```python
fps_cols = [c for c in df.columns if 'FPS' in c.upper()]

# Preencher valores nulos nas colunas de FPS
for c in fps_cols:
    df[c] = df[c].fillna(df[c].mean())

# Converter preço para numérico
df['price'] = pd.to_numeric(df['price'], errors='coerce')
df['price'] = df['price'].fillna(df['price'].mean())

df.head()
```

## Criando Métricas Importantes

```python
df['fps_medio'] = df[fps_cols].mean(axis=1)
df['custo_beneficio'] = df['fps_medio'] / df['price']
df.head()
```

## Visualização Inicial (Opcional)

```python
plt.figure(figsize=(10,4))
plt.hist(df['fps_medio'])
plt.title('Distribuição do FPS Médio')
plt.show()
```
